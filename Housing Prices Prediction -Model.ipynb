## Let's look at the histogram of target variable(SalePrice)

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
sns.distplot(train['SalePrice'], kde=False)
```

### The histogram reveals that the distribution of SalePrice is skewed. We need to unskew the variable before feeding it to our model.

```python
train.SalePrice.skew() #The skew of SalePrice is 1.88
```

```python
from scipy.stats import skew
from scipy.special import boxcox1p
from scipy.stats import boxcox_normmax
from scipy.special import inv_boxcox1p
```

```python
sale = train.SalePrice
```

```python
train['SalePrice'] = boxcox1p(train['SalePrice'], boxcox_normmax(1+train.SalePrice))
```

```python
normmax_saleprice = boxcox_normmax(1+sale)
```

```python
plt.figure(figsize=(10,6))
sns.distplot(train.SalePrice, kde=False)
```

### Now the variable looks normally distributed.

```python
train.SalePrice.skew() #The skew of the variable has reduced to 0.029
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='MSSubClass', y='SalePrice', data=train)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(y='Neighborhood', x='SalePrice', data=train)
```

### Neighborhood seems to be a good predictor of the sale price of the house. Houses which sell for maximum amount have neighborhoods like like NoRidge, NridgHt and StoneBr

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='SaleType', y='SalePrice', data=train)
```

```python
all_data = pd.concat([train, test]).reset_index(drop=True)
```

```python
all_data.shape
```

```python
all_data.head()
```

```python
missing = pd.DataFrame((100*all_data.isnull().sum()/len(all_data)).sort_values(ascending=False), columns=['Missing Values (%)'])
```

```python
missing = missing[missing>0].dropna()
```

```python
missing
```

```python
redundant_variables = [col for col in all_data.select_dtypes(exclude=['number']) if 1 - sum(all_data[col] == (all_data[col]).mode()[0])/len(all_data) < 0.03]
```

```python
redundant_variables
```

```python
all_data.drop(redundant_variables, axis=1, inplace=True)
```

## Let's fill the missing values in our dataset.

```python
for col in ('PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType', 'GarageQual', 'GarageFinish', 'GarageCond',
           'BsmtExposure', 'BsmtCond', 'BsmtQual', 'BsmtFinType2', 'BsmtFinType1', 'MasVnrType'):
    all_data[col].fillna('None', inplace=True)
```

```python
for col in ('GarageYrBlt', 'MasVnrArea', 'BsmtFullBath', 'BsmtHalfBath', 'GarageArea', 'BsmtFinSF2', 'BsmtFinSF1', 'TotalBsmtSF',
            'BsmtUnfSF', 'GarageCars'):
    all_data[col].fillna(0, inplace=True)
```

```python
map_LotFrontage = train.groupby('Neighborhood')['LotFrontage'].median().to_dict()
```

```python
all_data['LotFrontage'] = all_data['LotFrontage'].fillna(all_data['Neighborhood'].map(map_LotFrontage))
```

```python
for col in ('MSZoning', 'Functional', 'Electrical', 'Exterior1st', 'KitchenQual', 'Exterior2nd', 'SaleType'):
    all_data[col].fillna(train[col].mode()[0], inplace=True)
```

```python
all_data.isnull().sum().sum() #All missing values are now filled.
```

```python
all_data.head()
```

```python
pd.options.display.max_rows=300
all_data.dtypes  #This shows the datatypes of all the attributes in our dataset
```

```python
for c in ('MSSubClass', 'YrSold', 'MoSold'): #Some ordinal attributes should really be categorical..
    all_data[c] = all_data[c].astype(str)
```

```python
id_all_data = all_data['Id']
salePrice_all_data = all_data['SalePrice']
```

```python
all_data.drop(['Id', 'SalePrice'], axis=1, inplace=True)
```

```python
numeric_feats = all_data.dtypes[all_data.dtypes!='object'].index
```

```python
from scipy.stats import skew
```

```python
skewed_feats = all_data[numeric_feats].apply(lambda x: x.skew()).sort_values(ascending=False)
```

```python
skewness = pd.DataFrame({'Skew' : skewed_feats})
```

```python
skewness = skewness[abs(skewness)>0.75].dropna()
```

```python
skewness
```

```python
skewed_features = skewness.index
```

```python
for x in skewed_features:
    all_data[x] = boxcox1p(all_data[x], boxcox_normmax(1+all_data[x]))
```

```python
all_data[skewed_features].skew()
```

```python
[col for col in all_data.select_dtypes(exclude=['number']) if 1 - sum(all_data[col] == (all_data[col]).mode()[0])/len(all_data) < 0.03]
```

```python
all_data.drop('PoolQC', axis=1, inplace=True)
```

```python
all_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in ('LotShape', 'Alley', 'LandContour', 'LandSlope', 'BldgType', 'HouseStyle', 'YearBuilt','YearRemodAdd',
            'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 
            'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageFinish', 'GarageQual', 'GarageCond',
            'PavedDrive', 'Fence', 'YrSold', 'MoSold'):
    label = LabelEncoder()
    label.fit(all_data[col].values)
    all_data[col] = label.transform(all_data[col].values)
```

```python
all_data = pd.get_dummies(all_data, drop_first=True)
```

```python
training = pd.concat([train['Id'], all_data[:len(train)], salePrice_all_data[:len(train)]], axis=1)
```

```python
training.head()
```

```python
temp = all_data[len(train):].reset_index(drop=True)
```

```python
test_id_dataframe = pd.DataFrame({'Id': test['Id']})
```

```python
testing = pd.concat([test_id_dataframe, temp], axis=1)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
xgboost = XGBRegressor(learning_rate=0.05, n_estimators=250, max_depth=6, colsample_bylevel=0.6, n_jobs=-1)
```

```python
cv_score = -cross_val_score(xgboost, training.drop(['Id', 'SalePrice'], axis=1), training['SalePrice'], cv=10, scoring='neg_mean_squared_error')
```

```python
rmse = np.sqrt(cv_score.mean())
```

```python
print('The Root mean squared error is:', np.round(rmse,4))
```

```python
xgboost.fit(training.drop(['Id', 'SalePrice'], axis=1), training['SalePrice'])
```

```python
preds = pd.DataFrame(inv_boxcox1p(xgboost.predict(testing.drop('Id', axis=1)), normmax_saleprice), columns=['SalePrice'], index=test.Id)
```
